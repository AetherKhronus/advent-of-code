#### Answer

```python
scores = {}
i = 1

for board in boards:
    
    scores[i] = (board , [])
    i = i + 1

bingo = ()

for num in input:
    
    if (bingo != ()):
        
        break
        
    for b in scores.keys():
        
        board = scores.get(b)[0]
        marked = scores.get(b)[1]
        indices = [ind for ind , x in enumerate(board) if x == num]
        
        for ind in indices:
            
            marked.append(ind)
        
        scores[b] = (board , marked)
        
        for bi in bingos:
            
            if (all(elem in marked for elem in bi)):
                
                bingo = (b , num)
                break
                
got_bingo = scores.get(bingo[0])
unmarked = list(got_bingo[0])

for m in got_bingo[1]:
    
    unmarked.remove(got_bingo[0][m])

for k in range(len(unmarked)):
    
    unmarked[k] = int(unmarked[k])
    
score = sum(unmarked) * int(bingo[1])

print()
print("Bingo in Board" , bingo[0] , "with number" , bingo[1])
print("Board:" , got_bingo)
print("Score:" , score)
```

### --- Part Two ---

On the other hand, it might be wise to try a different strategy: let the <ins>giant squid win</ins>.

You aren't sure how many bingo boards a giant squid could play at once, so rather than waste time counting its arms, the safe thing to do is to __figure out which board will win last__ and choose that one. That way, no matter which boards it picks, it will win for sure.

In the above example, the second board is the last to win, which happens after `13` is eventually called and its middle column is completely marked. If you were to keep playing until this point, the second board would have a sum of unmarked numbers equal to `148` for a final score of `148 * 13 =` __`1924`__.

Figure out which board will win last. __Once it wins, what would its final score be__?
