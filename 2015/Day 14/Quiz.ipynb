### --- Part One ---

This year is the Reindeer Olympics! Reindeer can fly at high speeds, but must rest occasionally to recover their energy. Santa would like to know which of his reindeer is fastest, and so he has them race.

Reindeer can only either be __flying__ (always at their top speed) or __resting__ (not moving at all), and always spend whole seconds in either state.

For example, suppose you have the following Reindeer:

- Comet can fly __`14 km/s for 10 seconds`__, but then must rest for __`127 seconds`__.
- Dancer can fly __`16 km/s for 11 seconds`__, but then must rest for __`162 seconds`__.

After one second, Comet has gone 14 km, while Dancer has gone 16 km. After ten seconds, Comet has gone 140 km, while Dancer has gone 160 km. On the eleventh second, Comet begins resting (staying at 140 km), and Dancer continues on for a total distance of 176 km. On the 12th second, both reindeer are resting. They continue to rest until the 138th second, when Comet flies for another ten seconds. On the 174th second, Dancer flies for another 11 seconds.

In this example, after the 1000th second, both reindeer are resting, and Comet is in the lead at __`1120`__ km (poor Dancer has only gotten `1056` km by that point). So, in this situation, Comet would win (if the race ended at 1000 seconds).

Given the descriptions of each reindeer (in your puzzle input), after exactly `2503` seconds, __what distance has the winning reindeer traveled__?

#### Test

```python
seconds = 1000

information = {}

for i in test_infos_1:
    
    words = i.split(" ")
    reindeer = words[0]
    speed = int(words[3])
    time = int(words[6])
    rest = int(words[13])
    information[reindeer] = {"s" : speed , "t" : time , "r" : rest}

distances = {}

for r in information.keys():
    
    distances[r] = (0 , "fly" , information.get(r).get("t"))

for sec in range(seconds):
    
    for r in distances.keys():
        
        speed = information.get(r).get("s")
        
        if (distances.get(r)[1] == "fly"):
            
            distances[r] = (distances.get(r)[0] + speed , "fly" , distances.get(r)[2] - 1)
        
        else:
            
            distances[r] = (distances.get(r)[0] , "rest" , distances.get(r)[2] - 1)
            
        
        if (distances.get(r)[2] == 0):
            
            if (distances.get(r)[1] == "fly"):
                
                distances[r] = (distances.get(r)[0] , "rest" , information.get(r).get("r"))
                
            else:
                
                distances[r] = (distances.get(r)[0] , "fly" , information.get(r).get("t"))
            

print("After" , sec + 1 , "seconds:")
best = 0
    
for r in distances.keys():
        
    best = max(best , distances.get(r)[0])
    print("  " , r , "went" , distances.get(r)[0] , "kms")

print()
print("Most Distance:" , best)
```

### --- Part Two ---

Seeing how reindeer move in bursts, Santa decides he's not pleased with the old scoring system.

Instead, at the end of each second, he awards one point to the reindeer currently in the lead. (If there are multiple reindeer tied for the lead, they each get one point.) He keeps the traditional 2503 second time limit, of course, as doing otherwise would be entirely ridiculous.

Given the example reindeer from above, after the first second, Dancer is in the lead and gets one point. He stays in the lead until several seconds into Comet's second burst: after the 140th second, Comet pulls into the lead and gets his first point. Of course, since Dancer had been in the lead for the 139 seconds before that, he has accumulated 139 points by the 140th second.

After the 1000th second, Dancer has accumulated __`689`__ points, while poor Comet, our old champion, only has `312`. So, with the new scoring system, Dancer would win (if the race ended at 1000 seconds).

Again given the descriptions of each reindeer (in your puzzle input), after exactly 2503 seconds, __how many points does the winning reindeer have__?

#### Test

```python
seconds = 1000

information = {}

for i in test_infos_1:
    
    words = i.split(" ")
    reindeer = words[0]
    speed = int(words[3])
    time = int(words[6])
    rest = int(words[13])
    information[reindeer] = {"s" : speed , "t" : time , "r" : rest}

distances = {}

for r in information.keys():
    
    distances[r] = (0 , "fly" , information.get(r).get("t") , 0)

for sec in range(1 , seconds + 1):
    
    for r in distances:
        
        speed = information.get(r).get("s")
        
        if (distances.get(r)[1] == "fly"):
            
            distances[r] = (distances.get(r)[0] + speed , "fly" , distances.get(r)[2] - 1 , distances.get(r)[3])
        
        else:
            
            distances[r] = (distances.get(r)[0] , "rest" , distances.get(r)[2] - 1 , distances.get(r)[3])
            
        
        if (distances.get(r)[2] == 0):
            
            if (distances.get(r)[1] == "fly"):
                
                distances[r] = (distances.get(r)[0] , "rest" , information.get(r).get("r") , distances.get(r)[3])
                
            else:
                
                distances[r] = (distances.get(r)[0] , "fly" , information.get(r).get("t") , distances.get(r)[3])
    
    best = 0
    for r in distances.keys():
        
        if(distances.get(r)[0] > best):
            
            best = distances.get(r)[0]
            
    raindeers = []
    
    for r in distances.keys():
        
        if(distances.get(r)[0] == best):
            
            raindeers.append(r)
    
    for rain in raindeers:
    
        distances[rain] = (distances.get(rain)[0] , distances.get(rain)[1] , distances.get(rain)[2] , distances.get(rain)[3] + 1)
            

print("After" , seconds , "seconds:")
best = 0
    
for r in distances:
        
    best = max(best , distances.get(r)[3])
    print("  " , r , "has" , distances.get(r)[3] , "points")

print()
print("Most Points:" , best)
```
